```python
import numpy as np
```

# FORWARD PASS

```python
X = np.array([0.1,0.4,0.8,0.2])
y = np.array([0,1,1,0])
# x add a column of ones
X = np.c_[X, np.ones((X.shape[0]))]
W = np.array([0.1,0.2])
 ## FORWARD PASS
total_loss = 0

#calculate net
net = np.dot(X,W)

# calculate the output - sigmoid function on net
out = np.divide(1,np.add(1,np.exp(-1*net)))

#calculate error
error = out - y
# calculate loss
total_loss =np.sum( 0.5*(error)**2)
```

```python
total_loss
```

# Backward Pass



```python

dloss_dout = error
dout_dnet = out*(1-out)
dnet_dw = X
d_loss_dw = np.dot(dloss_dout * dout_dnet , dnet_dw)
W += d_loss_dw
```
